```python
import cv2
import matplotlib.pyplot as plt
```

<h2>CLAHE Test</h2>

```python
img = cv2.imread('fire_520.png')
```

```python
"""
Testing the effect of CLAHE on the image

Reference
<https://docs.opencv.org/4.x/d5/daf/tutorial_py_histogram_equalization.html>
"""


# converting to LAB color space as CLAHE affects the luminoisity of the image by design of the histogram equilisation
# converting to greyscale loses color information,HSV or sth else also works
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

# create a CLAHE object.
clahe = cv2.createCLAHE(clipLimit=4, tileGridSize=[5, 5])
lab[:,:,0] = clahe.apply(lab[:,:,0]) #apply on L

# Converting image from LAB Color model back to BGR color space
enhanced_img = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
```

```python
#ploting difference
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 5),
                       sharex=True, sharey=True)

#plt.gray()

ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGRA2RGB))
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(cv2.cvtColor(enhanced_img, cv2.COLOR_BGRA2RGB))
ax[1].axis('off')
ax[1].set_title('Augmented')

fig.tight_layout()

plt.show()
```
